```python
class Board:

    def __init__(self):
        self.board = []
        self._create_board()

    def _create_board(self):

        self.board = [[""] * 50 for i in range(50)]

    def print_board(self):
        """
        Prints the current state of the board.
        """
        for i in range(len(self.board)):
            tmp_str = "|"
            for j in self.board[i]:
                if j is None or j == "":
                    tmp_str += "   |"
            print(tmp_str)

```

```python
board = Board()
```

```python
board.print_board()
```
